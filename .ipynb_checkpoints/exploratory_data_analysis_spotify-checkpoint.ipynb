## Exploratory Data Analysis of Best Songs on Spotify for every year (2000-2023)

## Information about dataset:

###### Genre - the genre of the track
###### Year - the release year of the recording. Note that due to vagaries of * releases, re-releases, re-issues and general madness, sometimes the release years are not what you'd expect.
##### Beats Per Minute (BPM) - The tempo of the song.
###### Energy - The energy of a song - the higher the value, the more energtic. song
###### Danceability - The higher the value, the easier it is to dance to this song.
###### Loudness (dB) - The higher the value, the louder the song.
###### Liveness - The higher the value, the more likely the song is a live recording.
###### Valence - The higher the value, the more positive mood for the song.
###### Length - The duration of the song.
###### Acousticness - The higher the value the more acoustic the song is.
###### Speechiness - The higher the value the more spoken word the song contains.
###### Popularity - The higher the value the more popular the song is.
###### Duration - The length of the song.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df = pd.read_csv("spotify_best_songs_2000_2023.csv",sep=';')
df.head()
data = df
```

```python
from sklearn.ensemble import RandomForestRegressor

# Separating independent variables and target variable
X = data.drop(['popularity', 'title', 'artist', 'top genre','year','duration'], axis=1)
y = data['popularity']

# Creating the Random Forest Regressor model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)

# Visualizing feature importance ranking
feature_importances = rf.feature_importances_

# Visualizing feature importance ranking
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(X.columns, feature_importances)
plt.xlabel('Feature Importance Score')
plt.ylabel('Variables')
plt.title('Feature Importance Ranking for Popularity')
plt.show()
```

```python
# Counting categorical variables
data['top genre'].value_counts()

# Visualizing the distribution of categorical variables
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
data['top genre'].value_counts().nlargest(15).plot(kind='bar')
plt.xlabel('Music Genres')
plt.ylabel('Frequency')
plt.title('Distribution of Music Types')
plt.show()
```

```python
dance_pop_data = data[data['top genre'] == 'dance pop']


X = dance_pop_data .drop(['popularity', 'title', 'artist', 'top genre','year','duration'], axis=1)
y = dance_pop_data ['popularity']

# Random Forest Regressor modelini oluturma
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)

# zellik nem sralamasn aln
feature_importances = rf.feature_importances_

# zellik nem sralamasn grselletirme
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(X.columns, feature_importances)
plt.xlabel('Feature Importance Score')
plt.ylabel('Variables')
plt.title('Feature Importance Ranking for Popularity in Dance Pop Genre')
plt.show()
```

```python
# Histogram of numeric variables
plt.figure(figsize=(10, 6))
data['bpm'].hist(bins=20)
plt.xlabel('BPM')
plt.ylabel('Frequency')
plt.title('BPM Distribution of Songs')
plt.show()
```

```python
# Calculating the correlation matrix between variables
correlation_matrix = data.corr()

# Visualizing the correlation matrix
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Variables')
plt.show()
```

```python
# Grouping by year and song title variables
popular_songs_by_year = data.groupby('year').apply(lambda x: x.nlargest(1, 'popularity')).reset_index(drop=True)

# Visualizing results
plt.figure(figsize=(12, 8))
plt.bar(popular_songs_by_year['year'], popular_songs_by_year['title'])
plt.xlabel('Yl')
plt.ylabel('Song Name')
plt.title('Top Songs for Popularity by Year')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
